```python
# Image processing library
import cv2 as cv

# Show image
import matplotlib.image as pltimg
import matplotlib.pyplot as plt

# BLAS lib
import numpy as np

%matplotlib inline
```

```python
img = pltimg.imread("images/waymo_car.jpg")
print(img.shape)
plt.imshow(img)
```

```python
gimg = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(gimg, cmap="gray")
```

```python
subimg = img[img.shape[0] // 2:, :img.shape[1] // 2, :]
print(subimg.shape)
plt.imshow(subimg)
```
