```python
# Blas lib
import numpy as np

# Image processing lib
import cv2 as cv

# Display lib
import matplotlib.pyplot as plt

%matplotlib inline
```

First get the image, process the colors channels and show the properties.

```python
flat_earth_red_screen = cv.imread("images/pizza_bluescreen.jpg")
pizza_blue_screen = cv.cvtColor(flat_earth_red_screen, cv.COLOR_BGR2RGB)

print(type(pizza_blue_screen), pizza_blue_screen.shape)
plt.imshow(pizza_blue_screen)
```

Create a mask to exclude the blue region

```python
lower_blue = (0, 0, 200)
upper_blue = (100, 100, 255)
mask_background = cv.inRange(pizza_blue_screen, lower_blue, upper_blue)
plt.imshow(mask_background, cmap="gray")
```

Apply the mask to remove the background.

```python
pizza_without_background = np.copy(pizza_blue_screen)
pizza_without_background[mask_background != 0] = [0, 0, 0]
plt.imshow(pizza_without_background)
```

Get the space background an dig a pizza hole.

```python
space_background = cv.imread("images/space_background.jpg")
space_background = cv.cvtColor(space_background, cv.COLOR_BGR2RGB)
space_background = space_background[0:mask_background.shape[0], 0:mask_background.shape[1], :]

space_with_pizza_hole = np.copy(space_background)

space_with_pizza_hole[mask_background == 0] = [0, 0, 0]
plt.imshow(space_with_pizza_hole)
```

Assemble the results !

```python
pizza_in_space = space_with_pizza_hole + pizza_without_background
plt.imshow(pizza_in_space)
```
